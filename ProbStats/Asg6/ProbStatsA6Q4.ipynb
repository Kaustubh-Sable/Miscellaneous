```
def calculate_SSE(data):
  n = len(data)
  x_bar = data[data.columns[0]].mean()
  y_bar = data[data.columns[1]].mean()
  data["TM"] = data["TIME"] * data["NVM"]
  sigma_xiyi = data[data.columns[2]].sum()
  data["Xsq"] = data["TIME"] * data["TIME"]
  sigma_xisq = data[data.columns[3]].sum()

  beta1 = (sigma_xiyi-n*x_bar*y_bar)/(sigma_xisq-n*x_bar*x_bar)

  beta0 = y_bar - beta1*x_bar

  data["Pred_NVM"] = (beta0+beta1*data["TIME"])

  data["Sq_Error"] = (data["NVM"] - data["Pred_NVM"])**2
  sse = data[data.columns[5]].sum()

  del data["TM"]
  del data["Xsq"]

  return sse
```

```
def plot_my_graph(data):
  plt.figure(figsize=(8, 5), dpi=70)
  x1 = data["TIME"]
  y1 = data["NVM"]
  plt.plot(x1, y1)

  x2 = data["TIME"]
  y2 = data["Pred_NVM"]
  plt.plot(x2, y2)
  plt.savefig("plot.png")
```
